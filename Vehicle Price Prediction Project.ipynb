```python
df['bore']=pd.to_numeric(df['bore'],errors='coerce') # coerce means any errors in row that row will dont's stop we can convert remaining rows
df['bore'].describe()
```

```python
df['price']=pd.to_numeric(df['price'],errors='coerce') # coerce means any errors in row that row will dont's stop we can convert remaining rows
df['price'].describe()
```

```python
#lets us removed unvanted columns which are  not useful
df=df.drop('normalized-losses',axis=1)	
df.head()
```

```python
df.columns
```

```python
df['make']
```

```python
cylinders_dict={ 'two':2, 'three':3, 'four':4, 'five':5, 'six':6, 'eight':8, 'twelve':12 } 
df['num-of-cylinders'].replace(cylinders_dict,inplace = True)
df['num-of-cylinders'].head()
```

```python
a = {'1bbl':1, '2bbl':2,
       '4bbl':4, 'idi':5, 'mfi':6,
       'mpfi':7, 'spdi':8, 'spfi':9}
df['fuel-system'].replace(a,inplace = True)
b= {'dohc':1, 'dohcv':2, 'l':3,
       'ohc':4, 'ohcf':5, 'ohcv':6,
       'rotor':7}
df['engine-type'].replace(b,inplace = True)
c = {'front':1, 'rear':2} 
df['engine-location'].replace(c,inplace = True)
d = {'4wd':1, 'fwd':2,'rwd':3}
df['drive-wheels'].replace(d,inplace = True)

e= {'alfa-romero':1, 'audi':2, 'bmw':3,
       'chevrolet':4, 'dodge':5, 'honda':6, 'isuzu':7,
       'jaguar':8, 'mazda':9, 'mercedes-benz':10, 'mercury':11,
       'mitsubishi':12, 'nissan':13, 'peugot':14, 'plymouth':15,
       'porsche':16, 'renault':17, 'saab':18, 'subaru':19,
       'toyota':20, 'volkswagen':21, 'volvo':22,
}   
df['make'].replace(e,inplace = True)
f = {'convertible':1,
       'hardtop':2, 'hatchback':3, 'sedan':4,
           'wagon':5}
df['body-style'].replace(f,inplace = True)
g = {'four':4, 'two':2}
df['num-of-doors'].replace(g,inplace = True)

h = {'std':0, 'turbo':1}
df['aspiration'].replace(h,inplace = True)
df.head(10)
```

```python
i = {'gas':0, 'diesel':1}
df['fuel-type'].replace(i,inplace = True)
df.head(10)
```

```python
df.isnull()
```

```python
df.isna().sum()
```

```python
df.drop('horsepower',axis=1,inplace=True)
df.drop('curb-weight',axis=1,inplace=True)
df.drop('peak-rpm',axis=1,inplace=True)
#df.drop('length',axis=1,inplace=True)
```

```python
df=df.dropna()
```

# step3:train split test

```python
from sklearn.model_selection import train_test_split

x=df.drop('price',axis=1)
y=df['price']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# step4:train the model

```python
from sklearn.linear_model import LinearRegression
linear_model=LinearRegression()
linear_model.fit(x_train,y_train)
```

```python
linear_model.score(x_train,y_train)
```

```python
linear_model.coef_
```

```python
predictors=x_train.columns # see the weight associated with perticuler feature
coef=pd.Series(linear_model.coef_, predictors).sort_values()
print(coef)
```

# step5:test algorithm

```python
#lets predict using linear regression model
y_predict =linear_model.predict(x_test)
```
